```python
import numpy as np 
import pandas as pd 


df = pd.read_csv('/kaggle/input/shoes-logistic-regression/shoe - shoe.csv')

df.head(5)
```

```python
df['class(y)'] = df['class(y)'].map({'Female': 1, 'Male': 0})

df.head(5)
```

```python
y=df.drop(['size(cm)'], axis=1)

x=df.drop(['class(y)'], axis=1)

x.head(5)
```

```python
y.head(5)
```

```python
from sklearn.linear_model import LogisticRegression

clf= LogisticRegression()

clf.fit(x, y)

clf.predict(x)
```

```python
Y_predict = clf.predict_proba(x)

Y_predict
```
